# Overview
- Data는 항상 머신러닝 알고리즘 학습에 필요한 형태로 제공되지 않는다.
  - transform을 통해 데이터를 조작하고 학습에 적합하게 만든다.
- torchvision.transform 모듈은 주로 사용하는 몇가지 transform을 제공한다.
## FashionMNIST
- feature : PIL image 형식
- label : integer
- 학습을 위해서는 normalized tensor feature, encoded label이 필요하다.
  - ToTensor, Lambda 사용해서 transformation
  - ToTensor() : PIL image, Numpy ndarray를 FloatTensor로 변환하고 이미지 픽셀 Intensity 값을 \[0, 1] 범위로 비례하여 scale 한다.
  - Lambda : 사용자 정의 lambda 함수를 적용한다.
    - 정수를 원핫으로 바꾸는 함수를 정의
    - 먼저 label 크기 10 짜리 zero tensor를 만든다.
    - scatter_를 호출해 주어진 정답 y에 해당하는 인덱스(value=1)에 할당한다.
    - target_transform = Lambda(lambda y: torch.zeros(
    10, dtype=torch.float).scatter_(dim=0, index=torch.tensor(y), value=1))

```python
import torch
```

```python
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(
    root = "data",
    train = True,
    download = True,
    transform=ToTensor(),
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
```
