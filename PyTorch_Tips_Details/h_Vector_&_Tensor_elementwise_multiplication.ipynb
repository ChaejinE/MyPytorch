- element를 5개 가지는 주황색 vector와 (5, 3, 3)의 tensor를 element-wise multiplication하는 연산이다.

```python
A = torch.ones(5, 3, 3)
A # Tensor
```

```python
v = torch.arange(1, 5)
v # Vector
```

- 채널 방향으로 element-wise multiplication 하면 Tensor는 Channel 방향을 1, 2, 3, 4, 5를 가지게 된다.
- 다음 절차를 따르는 방법을 많이 사용한다.
  - 1. view를 이용해 벡터를 텐서와 같은 shape으로 맞추어 텐서로 만든다.
  - 2. 두 텐서를 곱한다.

```python
A = torch.ones(5, 3, 3)
print("A.shape : ", A.shape)
A
```

```python
v = torch.arange(0, 5)
print("v.shape : ", v.shape)
v
```

```python
v_tensor = v.view(v.size()[0], 1, 1)
print(v_tensor.size())
v_tensor
```

```python
result = v_tensor * A
print("result.shape : ", result.shape)
result
```
